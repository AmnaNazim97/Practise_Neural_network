```python
x= np.arange(1,8)
plt.plot(x, x*3 , label='Fast')
plt.plot(x,x**2 , label='Moderate')
plt.plot(x , x/2 , label = 'Slow')
plt.xlabel('Speed')
plt.ylabel('Time')
plt.grid=True
plt.legend(loc= 'upper left')
plt.savefig('abc.png')
plt.show()
```
